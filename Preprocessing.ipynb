```python
def check_float_convert(x):
    tok = x.split("-")
    if (len(tok))==2:
        return (float(tok[0]) + float(tok[1]))/2
    else:
        try:
            return float(x)
        except:
            return np.NAN
```

```python
data["total_sqft_processed"] = data["total_sqft"].apply(check_float_convert)
```
